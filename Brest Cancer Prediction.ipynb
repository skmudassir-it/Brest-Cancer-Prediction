### Model Traning

```python
X = df_cancer.iloc[:,:-1]
y = df_cancer.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=57)
```

```python
print('X train shape is: ',X_train.shape)
print('X test shape is: ',X_test.shape)
print('y train shape is: ',y_train.shape)
print('y test shape is: ',y_test.shape)
```

```python
from sklearn.svm import SVC 
from sklearn.metrics import classification_report, confusion_matrix

svc_model = SVC()
svc_model.fit(X_train, y_train)
```

### Model Evaluation

```python
y_predict = svc_model.predict(X_test)
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

### changing random state in splitting

Identify type1 and type2 error index value from confussion matrix

```python
print(cm)
print(cm[0][1])
print(cm[1][0])
```

Extracting all values which have type 2error as 0 and appending type 1 and type 2 valuesto a list

```python
l1 = []
li = []
for i in range(101):
    print("\n for random stateis ",i,"\n")
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=i)
    svc_model = SVC()
    svc_model.fit(X_train, y_train)
    y_predict = svc_model.predict(X_test)
    cm = confusion_matrix(y_test, y_predict)
    #print('confusion matrix before normalization\n',cm)
    min_train = X_train.min()
    range_train = (X_train - min_train).max()
    X_train_scaled = (X_train - min_train)/range_train
    min_test = X_test.min()
    range_test = (X_test - min_test).max()
    X_test_scaled = (X_test - min_test)/range_test
    svc_model = SVC()
    svc_model.fit(X_train_scaled, y_train)
    y_predict = svc_model.predict(X_test_scaled)
    cm = confusion_matrix(y_test, y_predict)
    #print('confusion matrix after normalization\n',cm)
    param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['rbf']} 
    grid = GridSearchCV(SVC(),param_grid, refit=True, verbose=4)
    grid.fit(X_train_scaled, y_train)
    grid_pred = grid.predict(X_test_scaled)
    cm = confusion_matrix(y_test, grid_pred)
    if cm[1][0] == 0:
        li.append(i)
        l1.append(cm[0][1])
```

```python
min(l1)
```

```python
print('least value in type1 error is ',min(l1))
```

```python
print('random state ',li,'\ntype1 error ',l1)
```
